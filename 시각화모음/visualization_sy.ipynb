```python
from matplotlib import font_manager, rc
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# 날씨 csv 파일 로드
month_df = pd.read_excel('data/2016_2022_usage_2.xlsx', engine='openpyxl')
month_df
```

```python
title='2016~2022 월별 전력 사용량'
ylabel='월별 사용량(kWh)'
xlabel='사용 월'

ax = month_df.plot.line(x='month',y=[2016,2017,2018,2019,2020,2021,2022],figsize=(16,3),lw=2,title=title);
ax.set(xlabel=xlabel, ylabel=ylabel)
ax.autoscale(axis='x', tight=True)
ax.legend(loc=0);
```

```python
title='2016~2022 월별 전력 사용량'
ylabel='월별 사용량'
xlabel='사용 월'

month_df = month_df.set_index('month')
ax = month_df.plot.bar(figsize=(16,3),title=title);
ax.set(xlabel=xlabel, ylabel=ylabel)
ax.legend(loc=0);
```

```python
title='2016~2022 월별 전력 사용량'
ylabel='월별 사용량'
xlabel='사용 월'

ax = month_df.plot.bar(stacked=True, title=title)
ax.set(xlabel=xlabel, ylabel=ylabel)
ax.legend(loc=4);
```

```python
title='2016~2022 월별 전력 사용량'
ylabel='월별 사용량'
xlabel='사용 월'

ax = month_df.plot.barh()
ax.set(xlabel=xlabel, ylabel=ylabel)
ax.legend(loc=4);
```

```python
title='2016~2022 월별 전력 사용량'
ylabel='월별 사용량'
xlabel='사용 월'

ax = month_df.plot.area(alpha=0.6, title=title)
ax.set(xlabel=xlabel, ylabel=ylabel)
ax.legend(loc=4);
```
