```python
import pandas as pd
```

칼럼명을 '2016년 1월'  '2016년 2월' 처럼 정제

```python
usage_2016 = pd.read_excel('전력사용량_16~22/2016.xlsx')
usage_2017 = pd.read_excel('전력사용량_16~22/2017.xlsx')
usage_2018 = pd.read_excel('전력사용량_16~22/2018.xlsx')
usage_2019 = pd.read_excel('전력사용량_16~22/2019.xlsx')
usage_2020 = pd.read_excel('전력사용량_16~22/2020.xlsx')
usage_2021 = pd.read_excel('전력사용량_16~22/2021.xlsx')
usage_2022 = pd.read_excel('전력사용량_16~22/2022.xlsx')
```

```python
usage_2016.columns = usage_2016.iloc[1]
usage_2017.columns = usage_2017.iloc[1]
usage_2018.columns = usage_2018.iloc[1]
usage_2019.columns = usage_2019.iloc[1]
usage_2020.columns = usage_2020.iloc[1]
usage_2021.columns = usage_2021.iloc[1]
usage_2022.columns = usage_2022.iloc[1]

# Remove the row with index 2 (line 3) from the DataFrame
usage_2016 = usage_2016.drop(0)
usage_2016 = usage_2016.drop(1)

usage_2017 = usage_2017.drop(0)
usage_2017 = usage_2017.drop(1)

usage_2018 = usage_2018.drop(0)
usage_2018 = usage_2018.drop(1)

usage_2019 = usage_2019.drop(0)
usage_2019 = usage_2019.drop(1)

usage_2020 = usage_2020.drop(0)
usage_2020 = usage_2020.drop(1)

usage_2021 = usage_2021.drop(0)
usage_2021 = usage_2021.drop(1)

usage_2022 = usage_2022.drop(0)
usage_2022 = usage_2022.drop(1)

# Reset the index
usage_2016 = usage_2016.reset_index(drop=True)
usage_2017 = usage_2017.reset_index(drop=True)
usage_2018 = usage_2018.reset_index(drop=True)
usage_2019 = usage_2019.reset_index(drop=True)
usage_2020 = usage_2020.reset_index(drop=True)
usage_2021 = usage_2021.reset_index(drop=True)
usage_2022 = usage_2022.reset_index(drop=True)
```

```python
usage_2016_T = usage_2016[usage_2016['시도'] == '광주광역시']
usage_2017_T = usage_2017[usage_2017['시도'] == '광주광역시']
usage_2018_T = usage_2018[usage_2018['시도'] == '광주광역시']
usage_2019_T = usage_2019[usage_2019['시도'] == '광주광역시']
usage_2020_T = usage_2020[usage_2020['시도'] == '광주광역시']
usage_2021_T = usage_2021[usage_2021['시도'] == '광주광역시']
usage_2022_T = usage_2022[usage_2022['시도'] == '광주광역시']
```

```python
usage_2016_df = usage_2016_T[usage_2016_T['계약종별'] == '합 계']
usage_2016_df = usage_2016_df.drop(['연도', '시도', '시군구', '계약종별'], axis = True)
usage_2016_df = usage_2016_df.reset_index(drop = True)
usage_2016_df
```
