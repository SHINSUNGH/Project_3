```python
# 필요 라이브러리 로드
from matplotlib import font_manager, rc
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import platform
# 정규식
import re
```

```python
elec_df = pd.read_excel('data/광주_3개년_월별_ 전력수요량.xlsx', engine='openpyxl')
```

```python
elec_df
```

```python
elec_df.drop(columns=['지역', '시군구', '용도별'], inplace=True)
```

```python
elec_df
```

```python
elec_df = pd.melt(elec_df, id_vars = ['연도'], var_name='월', value_name='kw')
```

```python
elec_df = elec_df.sort_values(by=['연도', '월']).reset_index(drop=True)
```

```python
elec_df
```

```python
elec_df = elec_df.rename(columns= {'연도' : 'year'})
```

```python
elec_df
```

```python
elec_df['month'] = elec_df['월'].apply(
    lambda x: int(re.findall('\d+',x.replace('월', ''))[0]) if pd.notnull(x) else x)
```

```python
elec_df.drop(columns='월', inplace=True)
```

```python
elec_df
```

```python
elec_df = elec_df[['year', 'month', 'kw']]
elec_df.sort_values(by=['year', 'month'], inplace=True)
elec_df
```
