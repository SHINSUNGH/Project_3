```python
#라이브러리 로드 
import matplotlib
import matplotlib. pyplot as pd
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
#CSV파일 로드 
pd.read_csv('data/광주_2021_2022_일별.csv',encoding='cp949')
weatherperd_df=pd.read_csv('data/광주_2021_2022_일별.csv',encoding='cp949')
weatherperd_df.head(3)
```

```python
# 필요없는 '지점','지점명' columns삭제
weatherperd_df.drop(columns='지점',inplace=True)
weatherperd_df.drop(columns='지점명',inplace=True)
weatherperd_df.tail(2)
```

```python
# columns이름 변경
weatherperd_df = weatherperd_df.rename(columns = {'일시' : 'date',
                                                           '평균기온(°C)' : 'temp',
                                                           '일강수량(mm)':'rainfall_total',             
                                                           '평균 상대습도(%)' : 'Humidity_mean',
                                                           '합계 일사량(MJ/m2)' :'solar',
                                                           '평균 전운량(1/10)' : 'cloud'})
weatherperd_df.tail(2)
```

```python
# DataFrame 정보 확인
weatherperd_df.info()
```

```python
#object 칼럼(date) ->year, month,day로 분리하여 칼럼 추가 
import re
weatherperd_df['year'] = weatherperd_df['date'].apply(
    lambda x: int(re.findall('\d+',x.split('-')[0])[0]) if pd.notnull(x) else x)

weatherperd_df['month'] = weatherperd_df['date'].apply(
    lambda x: int(re.findall('\d+',x.split('-')[1])[0]) if pd.notnull(x) else x)

weatherperd_df['day'] = weatherperd_df['date'].apply(
    lambda x: int(re.findall('\d+',x.split('-')[2])[0]) if pd.notnull(x) else x)

weatherperd_df
```

```python
#결측치교체 
weatherperd_df = weatherperd_df.fillna(0)
weatherperd_df.isnull().sum()
```

```python
# 칼럼명 순서 리스트로 변환
cols = list(weatherperd_df.columns)
cols
```

```python
# 칼럼명을 리스트로 순서 변경 
weatherperd_df = weatherperd_df[cols[5:] + cols[:5]]
weatherperd_df
```
